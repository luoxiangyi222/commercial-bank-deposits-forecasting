# Helper functions

```python
# reformat
date_reformat_f = lambda s: arrow.get(str(s), "YYYY MMM ").format("YYYY-MM")

year_reformat_f = lambda s: int(s)
```

# monthly - CPI

```python
original_df = pd.read_csv(
    "./raw-dataset\monthly_Consumer Price Index (CPI), 2019 As Base Year.csv"
)
df = original_df[9:11].copy().T
df.columns = df.iloc[0].values
df = df.iloc[1:]

# display original df
display(df.head())


df[["Data Series"]] = df[["Data Series"]].applymap(date_reformat_f)
df = df.sort_values(by="Data Series").reset_index(drop=True)
df = df.rename(columns={"All Items": "CPI_all_items_base_2019"})

display(df.head())

# save cleaned data
df.to_csv(
    "./clean-dataset/cln_monthly_Consumer Price Index (CPI), 2019 As Base Year.csv",
    index=False,
)
```

## monthly - compounded SORA

```python
from pandas.tseries.offsets import MonthEnd

filename = "daily_Domestic Interest Rates.csv"


original_df = pd.read_csv(f"./raw-dataset/{filename}", skiprows=4)

df = original_df.copy().iloc[:-9, 3:].dropna()
df = df[df["SORA Publication Date"] != "SORA Publication Date"]

# display(df.head())


df["Year Month"] = df["SORA Publication Date"].apply(
    lambda s: arrow.get(str(s), "DD MMM YYYY").format("YYYY-MM")
)

df["SORA Publication Date"] = df["SORA Publication Date"].apply(
    lambda s: arrow.get(str(s), "DD MMM YYYY").format("YYYY-MM-DD")
)


df = df.groupby(by=["Year Month"]).last().reset_index()

# remain month 3, 6, 9, 12
df = df[df["Year Month"].str.slice(-2).astype("int32").isin([3, 6, 9, 12])]


def month_to_quarter(s):
    word_ls = s.split("-")
    year_str, month_int = word_ls[0], int(word_ls[1])

    map_dict = {3: "1Q", 6: "2Q", 9: "3Q", 12: "4Q"}

    month_str = map_dict[month_int]

    return f"{year_str} {month_str}"


df["Year Month"] = df["Year Month"].apply(month_to_quarter)

df = df.rename(columns={"Year Month": "Data Series"})

df = df.drop(columns=["SORA Publication Date"])


for c in df.columns[1:]:
    df[c].mask(df[c] == "-", np.nan, inplace=True)

df[df.columns[1:]] = df[df.columns[1:]].astype("float64")

display(df.head())

# save cleaned data
df.to_csv(
    f"./clean-dataset/cln_quarterly_Domestic Interest Rates.csv", index=False,
)
```
