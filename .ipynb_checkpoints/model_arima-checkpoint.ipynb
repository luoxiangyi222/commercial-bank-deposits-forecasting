# ARIMA family: need to select appropriate hyper-parameters

- [Time Series Analysis: Identifying AR and MA using ACF and PACF Plots](https://towardsdatascience.com/identifying-ar-and-ma-terms-using-acf-and-pacf-plots-in-time-series-forecasting-ccb9fd073db8)

```python

import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

from src.data_prepare.data_loader import load_data
```

```python
Y_COL_NAME = "total_deposits_from_non-bank_customer"

## Load original dataframe

file_names = sorted(glob.glob("analyze-dataset/merged_data_*.csv"))

original_filename = file_names[-1]

# use cyclical encoding for quarter number


# to avoid data leakage for model evaluation
insample_df, outsample_df, full_df = load_data(original_filename, insample_end_year=2017)

# to avoid SettingWithCopyWarning
insample_df = insample_df.copy()
outsample_df = outsample_df.copy()


# cycling encoding for quarter number
insample_df["Qrt"] = (
    insample_df["Qrt"]
    .apply(lambda x: np.sin(x * (1 / 4) * (2 * np.pi)))
    .astype("int32")
)

INSAMPLE_ROWS = insample_df.shape[0]
OUTSAMPLE_ROWS = outsample_df.shape[0]

# y difference 
full_df['y_change'] = (full_df[Y_COL_NAME] - full_df[Y_COL_NAME].shift(1)) / full_df[Y_COL_NAME].shift(1)

insample_df = full_df.iloc[:INSAMPLE_ROWS].dropna()
outsample_df = full_df.iloc[-OUTSAMPLE_ROWS:]


print("insample: ", insample_df.shape)
print("outsample: ", outsample_df.shape)
```

### Testing For Stationarity  
- Ho: It is non stationary
- H1: It is stationary

```python

from statsmodels.tsa.stattools import adfuller
test_result=adfuller(insample_df['y_change'])


def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("P value is less than 0.05 that means we can reject the null hypothesis(Ho). Therefore we can conclude that data has no unit root and is stationary")
    else:
        print("Weak evidence against null hypothesis that means time series has a unit root which indicates that it is non-stationary ")
        
        
adfuller_test(insample_df['y_change'])
```

```python
# calculate diff of y

plt.figure(figsize=(20, 6))
plt.title(' Commercial banks deposits quarter-on-quarter Percentage change.')
plt.xticks(rotation=90)
plt.plot(full_df['Data Series'], full_df['y_change'])
plt.axhline(y=0.0, color="r", linestyle="--", alpha=0.7)
plt.show()
```

## According the ACF and PACF plots above, we need to try MA(1), AR(1), ARMA(1,1), ARIMA(1,1,1)

- Autocorrelation plot

```python
plt.figure()
plt.title('The Autocorrelation Plot')
plt.acorr(insample_df['y_change'], maxlags = 15)  
plt.axhline(y = 0.8, color = 'r', linestyle = '--', alpha=0.7)
print("The Autocorreleation plot for the data is:")
plt.grid(True)
plt.show()
```

- ACF Plot

```python
plt.figure()
plot_acf(insample_df['y_change'])
plt.show()
```

- PACF Plot

```python
plt.figure()
plot_pacf(insample_df[Y_COL_NAME], method='ywm')
plt.show()
```

## MA(1)

- [Forecasting in statsmodels](https://www.statsmodels.org/dev/examples/notebooks/generated/statespace_forecasting.html)

```python
insample_df['Data Series']
```

```python
insample_df.index = pd.period_range('1992Q2', '2017Q4', freq='Q')


arima_model = ARIMA(insample_df['y_change'], order=(0, 0, 1))
fitted_model = arima_model.fit()
fitted_model.summary()

# plot of residuals
residuals = pd.DataFrame(fitted_model.resid)
residuals.plot()
plt.show()

# density plot of residuals
residuals.plot(kind='kde')
plt.show()

# summary stats of residuals
# found that mean is nearly zero
print(residuals.describe())


plt.figure(figsize=(20, 6))
plt.xticks(rotation=90)
plt.plot(insample_df['Data Series'],insample_df['y_change'], label='real y')
plt.plot(insample_df['Data Series'],fitted_model.fittedvalues, label='fitted y')
plt.legend()
plt.show()



forecast_out = fitted_model.forecast(steps=OUTSAMPLE_ROWS)

plt.figure()
plt.xticks(rotation=90)
plt.plot(outsample_df['Data Series'], outsample_df['y_change'], label='real y')
plt.plot(outsample_df['Data Series'], forecast_out, label='fitted y')
plt.legend()
plt.show()
print(forecast_out)
```
